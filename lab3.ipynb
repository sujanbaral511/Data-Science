# TITLE: DATA VISUALIZATION USING HEALTH DATASET

```python
# Importing Python modules
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

# Load Dataset

```python
# Load Dataset
def load_health(file_path):
    return pd.read_csv(file_path)

df = load_health("./Health.csv")
df.head(7)
```

# Line Chart

```python
# Use first 20 data and sort them by height. 
subset = df.head(20).sort_values(by='heightCm')

# Extract sorted height and weight
height = subset['heightCm']
weight = subset['weightKg']

plt.figure(figsize=(10, 6))
# create a line chart, heightCm on x- axis, weightKg on y-axis
plt.plot(height, weight, color = 'blue', marker = 'o', linestyle = '-')

# add a title
plt.title("Weight Variation Across Different Heights")

# add a label to the x-axis and y-axis
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")

# Plot grid in graph
plt.grid(True)

# Display the plot 
plt.show()
```

# Findings:
-> The graph titled "Weight Variation Across Different Heights" shows a general increasing trend between height and weight.

-> However, there is also a lot of variation in weight at each height, indicating that other factors besides height play a significant role in determining weight.

# bar plot

```python
# Use first 10 data and sort them by weight
subset = df.head(10).sort_values(by='weightKg')

# Extract sorted weight and Heart Rate
age = subset['age']
weight = subset['weightKg']

# Plot the Bar Graph 
plt.bar(range(len(age)), weight)

# Add the title 
plt.title("Age vs Weight")

# Add the label in x-axis and y-axis
plt.ylabel("Age")
plt.xlabel("Weight")

plt.xticks(range(len(age)), age)
plt.tight_layout()  # Adjust layout to prevent overlap

# Display the plot
plt.show()
```

### Findings:
-> The graph shows positive correlation between age and weight, meaning as age increases, weight tends to increase as well.

-> However, there's a lot of variation among individuals of the same age, suggesting other factors besides age influence weight.

```python
df['sex'] = df['sex'].map({1:'Male', 0:'Female'})
sns.countplot(data=df, x='sex')
df['BMI'].value_counts()
```

Findings:
-> The graph shows a near-equal distribution of males and females in the dataset.

-> The bars for 'Male' and 'Female' are almost the same height, suggesting that the dataset is roughly balanced in terms of gender representation.

-> "df['BMI'].value_counts()" counts the number of people with certain BMI and prints it.

# Scatter Plot

```python
# Create a scatter plot
subset = df.head(100)
SysBP = subset['SysBP']
DiaBP = subset['DiaBP']
plt.scatter(SysBP, DiaBP)

# Add labels and a title
plt.xlabel('Systolic Blood Pressure')
plt.ylabel('Diastolic Blood Pressure')
plt.title('Scatter Plot of SysBP vs DiaBP')

# Show the plot
plt.show()
```

### Findings:
-> The scatter plot shows the relationship between systolic blood pressure (SysBP) and diastolic blood pressure (DiaBP) for a subset of 10 individuals.

-> The data points are somewhat scattered, suggesting that while there's a general trend, there's also variability in the relationship.
